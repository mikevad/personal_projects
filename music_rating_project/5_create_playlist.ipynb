```python
import numpy as np
import pandas as pd
```

```python
recommendations = pd.read_csv('first_recommendations.csv')
```

```python
recommendations.head()
```

```python
all_music = pd.read_csv('all_music.csv')
```

```python
all_music.head()
```

```python
playlist_df = pd.merge(all_music, recommendations, how = 'inner', left_on = ['Artist', 'Name'], right_on = ['artist_name', 'song_title'])
```

```python
playlist_df.shape
```

```python
playlist_df = playlist_df[playlist_df['recommendation'].notnull()]
```

```python
playlist_df.head()
```

```python
playlist_df = playlist_df.drop(columns = ['lyrics', 'song_title', 'artist_name', 'liked', 'cleaned_lyrics'])
```

```python
playlist_df.recommendation.value_counts(bins = 100).sort_index(ascending = False)
```

```python
playlist_df[playlist_df.recommendation == 1.0].shape
```

```python
export_df = playlist_df.sort_values(by = 'recommendation', ascending = False).iloc[100:200]
```

```python
export_df = export_df.drop(columns = 'recommendation')
```

```python
with open('recommedations_v0_1','w') as write_tsv:
    write_tsv.write(export_df.to_csv(sep='\t', index=False))
```
